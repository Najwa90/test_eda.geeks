```python
#There is  thing that take  attention , agnosis is our class label  "Unnamed: 2" feature includes NaN so we do not need it.
df.drop('Unnamed: 2',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Replace the original column values with the mapped values
df['Status'] = df['Status'].map({'Alive': 1, 'Dead': 0})
df['Progesterone Status'] = df['Progesterone Status'].map({'Positive': 1, 'Negative': 0})
df['Estrogen Status'] = df['Estrogen Status'].map({'Positive': 1, 'Negative': 0})
df['N Stage'] = df['N Stage'].map({'N1': 0, 'N2': 1,'N3': 2})
df['6th Stage'] = df['6th Stage'].map({'IIA': 0, 'IIB': 1,'IIIA': 2, 'IIIB': 3,'IIIC': 4})
df['Grade'] = df['Grade'].map({'Moderately differentiated Grade II':0, 'Poorly differentiated Grade III':1,'Well differentiated; Grade I':2, 'Undifferentiated; anaplastic; Grade IV':3})
df['A Stage'] = df['A Stage'].map({'Distant': 0, 'Regional': 1})
```

```python
df.head()
```

```python
#Step 6: Data Visualization
# Histogram of Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()


```

```python
# Bar plot of Status distribution
plt.figure(figsize=(10, 6))
sns.countplot(df['Status'])
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Status Distribution')
plt.show()

```

```python
# Scatter plot of Age vs. Survival Months
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Survival Months', data=df)
plt.xlabel('Age')
plt.ylabel('Survival Months')
plt.title('Age vs. Survival Months')
plt.show()


```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import seaborn as sns

# Generate a scatter plot matrix with the columns
cols = ['Status', 'Age', 'Tumor Size', 'Regional Node Examined', 'Reginol Node Positive', 'Survival Months']
sns.pairplot(data=df[cols], hue='Status', palette='rocket')


# Print the legend
print("Alive = 1")
print("Dead = 0")
```

```python
columns_to_compare = ['A Stage','Estrogen Status','Progesterone Status',  'N Stage', '6th Stage','Regional Node Examined', 'Reginol Node Positive', 'Tumor Size','Survival Months']

# Melt the dataframe to reshape the data for box plot
melted_data = pd.melt(df, id_vars='Status', value_vars=columns_to_compare, var_name='Column')

# Display the melted data
print(melted_data.head())
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Column', y='value', data=melted_data)
plt.xlabel('Column')
plt.ylabel('Value')
plt.title('Comparison of Columns')
plt.xticks(rotation=45)
plt.show()
```
